```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
male = 0
student = data.shape[0]
for i in data['sex']:
    if i == 'M':
        male += 1
female = student - male
labels = 'Male', 'Female'
sizes = [male, female]
explode = (0, 0.09)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
data1  = data.T
```

```python
d = np.zeros(data.shape)
```

```python
for i in range(0, data1.shape[1]):
    for j in range(0, data1.shape[0]):
        if type(data1[i][j]) == int:
            d[i][j] = data1[i][j]
        elif j == 0:
            if data1[i][j] == 'MS':
                d[i][j] = 1
            else:
                d[i][j] = 0
        elif j == 1:
            if data1[i][j] == 'M':
                d[i][j] = 1
            else:
                d[i][j] = 0
        elif j == 3:
            if data1[i][j] == 'U':
                d[i][j] = 1
            else:
                d[i][j] = 0
        elif j == 4:
            if data1[i][j] == 'GT3':
                d[i][j] = 1
            else:
                d[i][j] = 0
        elif j == 5:
            if data1[i][j] == 'T':
                d[i][j] = 1
            else:
                d[i][j] = 0
        elif j == 8 or j == 9:
            if data1[i][j] == 'teacher':
                d[i][j] = 0
            elif data1[i][j] == 'health':
                d[i][j] = 1
            elif data1[i][j] == 'services':
                d[i][j] = 2
            elif data1[i][j] == 'at_home':
                d[i][j] = 3
            elif data1[i][j] == 'other':
                d[i][j] = 4
        elif j == 10:
            if data1[i][j] == 'home':
                d[i][j] = 0
            elif data1[i][j] == 'reputation':
                d[i][j] = 1
            else:
                d[i][j] = 2
        elif j == 11:
            if data1[i][j] == 'mother':
                d[i][j] = 0
            elif data1[i][j] == 'father':
                d[i][j] = 1
            else:
                d[i][j] = 2
        else:
            if data1[i][j] == 'yes':
                d[i][j] = 1
            else:
                d[i][j] = 0
```

```python
def swap(a, b):
    for i in range(0, d.shape[1]):
        temp = d[a][i]
        d[a][i] = d[b][i]
        d[b][i] = temp
```

```python
import random
for i in range(0, 1000):
    a = random.randint(0, d.shape[0]-1)
    b = random.randint(0, d.shape[0]-1)
    if a != b:
        swap(a, b)
```

```python
from sklearn.linear_model import LinearRegression, Lasso 
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
import sklearn.tree
```

```python
X = d[:,0:d.shape[1]-1]
Y = np.zeros(d.shape[0])
for i in range(0, d.shape[0]):
    Y[i] = d[i][d.shape[1]-1]
```

```python
print("accuracy using linear regression:")
print(cross_val_score(LinearRegression(), X, Y, cv = 7).mean())
```

```python
print("accuracy using Lasso:")
print(cross_val_score(Lasso(), X, Y, cv = 7).mean())
```

```python
print("accuracy using Ridge:")
print(cross_val_score(Ridge(), X, Y, cv = 7).mean())
```

```python
model = Sequential()
model.add(Dense(20, input_dim = 32, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(5, activation = 'relu'))
model.add(Dense(1, activation = 'relu'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, validation_split=0.20, epochs=1000)
```
